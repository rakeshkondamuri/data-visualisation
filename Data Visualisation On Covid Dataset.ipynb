```python
df_Italy_2021 = df[(df.Country == 'Italy') & (df['Date_reported'] >= '2021-01-01') & (df['Date_reported']<='2021-01-15')]
```

```python
df_India_2021 = df[(df.Country == 'India') & (df['Date_reported'] >= '2021-01-01') & (df['Date_reported']<='2021-01-15')]
```

```python
import matplotlib.pyplot as plt
import numpy as np
world_date = df_India_2021.loc[(df_India_2021["Date_reported"] >= '2021-01-01') & (df_India_2021["Date_reported"] <= '2021-01-30')]
x=world_date["Date_reported"]
y=world_date["New_deaths"]
z=world_date["New_cases"]
x = np.arange(len(x))
fig, ax = plt.subplots()
graph1=plt.plot(x,y,color='red', linestyle='dotted', linewidth = 1,marker='o', markerfacecolor='red', markersize=2,label="deaths")
graph2=plt.plot(x,z,color='blue', linestyle='dotted', linewidth = 1,marker='*', markerfacecolor='blue', markersize=7,label="cases")
plt.xlabel("Feb 2021 (day's)", fontname="Gabriola", fontsize=18)
plt.ylabel("New Death's And Cases's In Feb 2021", fontname="Gabriola", fontsize=18)
plt.title("India's Covid Situation In February  2021",fontname='Franklin Gothic Medium', fontsize=18,y=1.10)
ax.set_xticks(x)
fig.tight_layout()
plt.legend()
sns.set(rc={'figure.figsize':(10,10)})
plt.grid(color = 'g', linestyle = '--', linewidth = 0.5)
plt.show()
```

### plot graph between italy and india's new cases 

```python
x = df_India_2021['Date_reported']
y = df_India_2021['New_cases']
z = df_Italy_2021['New_cases']
plt.plot(x,y,label='New Cases : India', color = 'g',marker = 'd')
plt.plot(x,z,label='New Cases : Italy',color = 'r',marker = 'd')
plt.title("Covid-19 New Cases")
plt.xlabel("Date Reported")
plt.ylabel("New Cases")
plt.legend()
sns.set(rc={'figure.figsize':(8,10)})
plt.xticks(rotation=90)
plt.show()
```

### pie chart for new cases in the india's dataset 

```python
temp=pd.DataFrame({"New_cases":df_India_2021['New_cases']})
count_New_cases=temp.value_counts()
#print(count_New_cases)
labels=[X[0] for X in count_New_cases.keys()]
bar,ax=plt.subplots(figsize=(10,10))
ax=plt.pie(x=count_New_cases,autopct="%.1f%%",labels = labels)
plt.title("New cases",fontsize=17) 
plt.show()
```
